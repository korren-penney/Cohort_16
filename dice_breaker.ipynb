```python
# The purpose of this program is to use the random module in python 
# to roll a dice for our icebreaker today. Korren is going to attempt
# to do this live, be prepared for typos.

# import the random library
import random

# declare our variable to determine the smallest and largest numbers
min = 1
max = 6

# decalre the variable for keeping track of if we need to roll again
# set var to yes to start our loop
roll_again = "yes"

# start the loop
while roll_again == 'yes':
    # print message to indicate the die is rolling
    print("Rolling the die...")
    
    # declare var for the die value and assign it a random number
    die_value = (random.randint(min,max))
    
    # print die value
    print("The die value is: ", die_value)
    
    # create the question if statement
    if die_value == 1:
        print("If you were invisible for a day, what would you do?")
    elif die_value == 2:
        print("If you could choose a super power, what would it be?")
    elif die_value == 3:
        print("What is the ggrossest food you have ever tried?")
    elif die_value == 4:
        print("Who is your favourite book or movie character?")
    elif die_value == 5:
        print("What is your favourite game to play?")
    else:
        print("What is one thing you are really good?")
        
    # Prompt the user to roll again
    roll_again = input("Roll again? yes/no?")
    
	# Ending the program and error handling 
    if roll_again == "no":
        print("you have selected no, program terminating")
    elif roll_again != "yes" or roll_again != "no":
        print("invalid entry, program terminating")      
    
print("This message shows the code is done!")
```
